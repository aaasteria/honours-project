```python
# -*- coding: utf-8 -*-
import rdflib
import re
import urllib2
import networkx as nx
import networkx as nx
import matplotlib.pyplot as plt
from itertools import count
g1 = rdflib.Graph()
g1.parse("data.rdf", format="xml")
G = nx.DiGraph()
i = 1
s=list()
p=list()
o=list()
ed1 = list()
ed2 = list()
while i:
    print "please input the three words that you want,Subject,Predicate,Object;ues ',' to split them,ues '' if you don't know:"
    try:
        Subject, Predicate, Object = input()
        data = ['', '', '']
        if Subject == '' and Predicate == '' and Object != '':  # only search Object
            q = "select ?Subject ?Predicate where {?Subject ?Predicate <http://baike.com/resource/" + Object + ">}"
            data[2] = Object.decode('utf-8')#.decode().encode('gbk')#.encode('gbk')
            x1 = g1.query(q)
            if len(list(x1)) > 20:
                leng = 20
            else:
                leng = len(list(x1))
            for i in range(leng):
                s1 = str(list(x1)[i]).decode('unicode_escape')
                txt = re.findall(r'resource/.*?\'', s1)
                data[0] = txt[0][9:-1]
                data[1] = txt[1][9:-1]
                print data[0], data[1], data[2]
                if data[0] not in G: 
                    s.append(data[0])
                    G.add_node(data[0])
                if data[1] not in G:
                    p.append(data[1])
                    G.add_node(data[1])
                if data[2] not in G:                    
                    o.append(data[2])
                    G.add_node(data[2])
                e1 = (data[0],data[1])
                ed1.append(e1)
                e2 = (data[1],data[2])
                ed2.append(e2)
            else:
                break
            i = 0
    except:
        pass
G.add_edges_from(ed1)
G.add_edges_from(ed2)
print len(G)
print len(s)
print len(p)
print len(o)

plt.figure(figsize=(19,12))
plt.axis('off')
pos = nx.spring_layout(G)
nx.draw_networkx_nodes(G, pos , nodelist=s, node_color='#FFB6C1', node_shape='8',with_labels=True,node_size=2000,font_weight='bold')
nx.draw_networkx_nodes(G, pos , nodelist=p, node_color='#DB7093', node_shape='o',with_labels=True,node_size=1000,font_weight='bold')
nx.draw_networkx_nodes(G, pos , nodelist=o, node_color='#DC143C', node_shape='p',with_labels=True,node_size=1500,font_weight='bold')
nx.draw_networkx_labels(G,pos , font_size=20,font_family='sans-serif')
nx.draw_networkx_edges(G, pos , edgelist=ed2,width=3, arrows=True,edge_color='#FA8072')
nx.draw_networkx_edges(G, pos , edgelist=ed1,width=3, arrows=True,edge_color='#BC8F8F')
plt.show()
    #so^>v<dph8
```
